```python
import pandas as pd
import numpy as np
```

# **11번가 장바구니 분석**

```python
eleven=pd.read_excel('2305_11번가장바구니data1_합본.xlsx')
```

```python
eleven.info()
```

# 데이터 전처리

```python
##결측치 처리하기

#결측치 있는 행 출력하기
eleven=eleven[eleven.isnull().any(axis=1)]
eleven=eleven.fillna(np.nan)
eleven
```

```python
##'주문일', '출고예정일', '운송장등록일자' 열의 데이터 타입을 datetime 형식으로 변환하기

eleven['주문일'] = pd.to_datetime(eleven['주문일'], format='%Y%m%d')
eleven['출고예정일'] = pd.to_datetime(eleven['출고예정일'], format='%Y%m%d')
eleven['운송장등록일자'] = pd.to_datetime(eleven['운송장등록일자'])
eleven.info()

#info 정보를 보면 datetime 형식으로 바뀐 것 확인 가능함.
```

```python
##'운송장등록일자'와 '주문일'열의 데이터를 계산하여 '운송장등록소요일' 열 추가하기
##'출고예정일'과 '주문일'열의 데이터를 계산하여 '출고소요일' 열 추가하기

eleven['운송장등록소요일'] = (eleven['운송장등록일자'] - eleven['주문일']).dt.days
eleven['출고소요일'] = (eleven['출고예정일'] - eleven['주문일']).dt.days
eleven
```

# 매출액/판매량 분석

```python
##상품유형별 groupby

eleven.groupby('상품유형')['매출액'].sum()
#단일 상품의 매출액이 더 높음.
```

```python
#상품명별 상품수량(판매량) 분석

eleven.groupby('상품명')['상품수량'].sum().sort_values(ascending=False).head(30)
```

```python
#평균 배송기간

print(eleven['운송장등록소요일'].mean())
print(eleven['출고소요일'].mean())
```

```python
#배송기간별 취소수량

print(eleven.groupby('운송장등록소요일')['취소수량'].sum())
print(eleven.groupby('출고소요일')['취소수량'].sum().sort_values(ascending=False))
#출고소요일(출고예정일-주문일)이 2일인 경우에 취소수량이 가장 많고, 그 다음은 5일임.
```

```python
#배송기간별 반품수량

print(eleven.groupby('운송장등록소요일')['반품수량'].sum().sort_values(ascending=False))
print(eleven.groupby('출고소요일')['반품수량'].sum().sort_values(ascending=False))
#운송장등록에 2일 소요되거나 출고가 5일 소요된 경우에 반품수량이 가장 많았음.
```

```python
#배송이 평균보다 오래걸리는 상품(출고소요일로 비교)

eleven[eleven['출고소요일'] > 2]
```

```python
#문제의 반품수량이 가장 많았던 기간의 상품들

eleven[(eleven['운송장등록소요일'] == 2) | (eleven['출고소요일'] == 5)]
```

```python
#취소수량이 가장 많았던 출고소요일이 2일 걸린 상품들

eleven[eleven['출고소요일'] == 2]
```
