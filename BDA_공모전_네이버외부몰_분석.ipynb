```python
import pandas as pd
import numpy as np
```

```python
naver=pd.read_excel('2305_네이버외부몰data2_합본.xlsx')
```

```python
naver
```

# 데이터 전처리

```python
##결측치 처리하기

#결측치 있는 행 출력하기
naver=naver[naver.isnull().any(axis=1)]
naver=naver.fillna(np.nan)
naver
```

```python
##'주문일', '출고예정일', '운송장등록일자' 열의 데이터 타입을 datetime 형식으로 변환하기

naver['주문일'] = pd.to_datetime(naver['주문일'], format='%Y%m%d')
naver['출고예정일'] = pd.to_datetime(naver['출고예정일'], format='%Y%m%d')
naver['운송장등록일자'] = pd.to_datetime(naver['운송장등록일자'])
naver.info()

#info 정보를 보면 datetime 형식으로 바뀐 것 확인 가능함.
```

```python
##'운송장등록일자'와 '주문일'열의 데이터를 계산하여 '운송장등록소요일' 열 추가하기
##'출고예정일'과 '주문일'열의 데이터를 계산하여 '출고소요일' 열 추가하기

naver['운송장등록소요일'] = (naver['운송장등록일자'] - naver['주문일']).dt.days
naver['출고소요일'] = (naver['출고예정일'] - naver['주문일']).dt.days
naver
```

# 매출액/판매량 분석

```python
##상품유형별 groupby

naver.groupby('상품유형')['매출액'].sum()
#단일 상품의 매출액이 가장 높음.
```

```python
##상품명별 상품수량(판매량) 분석

#상위 30개
naver.groupby('상품명')['상품수량'].sum().sort_values(ascending=False).head(30)
```

```python
#하위 30개
naver.groupby('상품명')['상품수량'].sum().sort_values().head(30)
```

```python
##매출액순 비교

#상위 30개
print(naver.groupby('상품명')['매출액'].sum().sort_values(ascending=False).head(30))
#햇반 유기농쌀밥 210Gx12개의 매출액이 가장 높음.
print(naver.groupby('상품명').size().sort_values(ascending=False).head(10))
```

```python
#하위 30개
print(naver.groupby('상품명')['매출액'].sum().sort_values().head(30))
```

```python
#비비고 제품만 따로 비교하기

naver_bibigo = naver[naver['상품명'].str.contains('비비고')]
print(naver_bibigo.groupby('상품명')['매출액'].sum().sort_values(ascending=False).head(10))
print(naver_bibigo.groupby('상품명').size().sort_values(ascending=False).head(10))

#비비고 한우사골곰탕 500Gx12개가 매출액 가장 많음.
#비비고 왕교자 1.05KGx2개+김치왕교자 1.05KGx2개는 매출액과 판매건수 모두 많은 편임.
```

```python
#비비고 제품의 상품유형별 매출액

naver_bibigo.groupby('상품유형')['매출액'].sum()
#비비고 제품은 전체 상품 추세와 다르게 복합 상품의 매출액이 가장 많으며, 단일 상품은 존재하지 않음.
```

```python
##백설 제품만 따로 비교하기

naver_baek = naver[naver['상품명'].str.contains('백설')]
print(naver_baek.groupby('상품명')['매출액'].sum().sort_values(ascending=False).head(10))
print(naver_baek.groupby('상품명').size().sort_values(ascending=False).head(10))

#백설 제품의 상품유형별 매출액

naver_baek.groupby('상품유형')['매출액'].sum()
```

```python
##다담(백설 하위 브랜드) 제품만 따로 비교하기

naver_dadam = naver[naver['상품명'].str.contains('다담')]
print(naver_dadam.groupby('상품명')['매출액'].sum().sort_values(ascending=False).head(10))
print(naver_dadam.groupby('상품명').size().sort_values(ascending=False).head(10))


#다담 제품의 상품유형별 매출액

naver_dadam.groupby('상품유형')['매출액'].sum()
```

```python
##햇반 제품만 따로 비교하기

naver_hat = naver[naver['상품명'].str.contains('햇반')]
print(naver_hat.groupby('상품명')['매출액'].sum().sort_values(ascending=False).head(10))
print(naver_hat.groupby('상품명').size().sort_values(ascending=False).head(10))
#타 브랜드에 비해 햇반의 매출과 판매량이 매우 많음.

#햇반 제품의 상품유형별 매출액

print(naver_hat.groupby('상품유형')['매출액'].sum())
#단일상품의 매출액이 매우 높고, 복합상품은 없음.
```

```python
#고메 제품만 따로 비교하기

naver_go = naver[naver['상품명'].str.contains('고메')]
print(naver_go.groupby('상품명')['매출액'].sum().sort_values(ascending=False).head(10))
print(naver_go.groupby('상품명').size().sort_values(ascending=False).head(10))

#고메 제품의 상품유형별 매출액

print(naver_go.groupby('상품유형')['매출액'].sum())
#고메 제품은 묶음 상품의 매출액이 가장 많음.
```

```python
#The더건강한 제품만 따로 비교하기

naver_the = naver[naver['상품명'].str.contains('더건강한')]
print(naver_the.groupby('상품명')['매출액'].sum().sort_values(ascending=False).head(10))
print(naver_the.groupby('상품명').size().sort_values(ascending=False).head(10))
#주문이 있었으나 모두 취소 혹은 반품되어 매출이 없음.
```

```python
#다시다 제품만 따로 비교하기

naver_dasida = naver[naver['상품명'].str.contains('다시다')]
print(naver_dasida.groupby('상품명')['매출액'].sum().sort_values(ascending=False).head(10))
print(naver_dasida.groupby('상품명').size().sort_values(ascending=False).head(10))
#주문이 있었으나 모두 취소 혹은 반품됨.
```

```python
#해찬들 제품만 따로 비교하기

naver_hae = naver[naver['상품명'].str.contains('해찬들')]
print(naver_hae.groupby('상품명')['매출액'].sum().sort_values(ascending=False).head(10))
print(naver_hae.groupby('상품명').size().sort_values(ascending=False).head(10))
#주문이 있었으나 모두 취소 혹은 반품됨.
```

```python
#스팸 제품만 따로 비교하기

naver_spam = naver[naver['상품명'].str.contains('스팸')]
print(naver_spam.groupby('상품명')['매출액'].sum().sort_values(ascending=True))
print(naver_spam.groupby('상품명').size().sort_values(ascending=False))
#스팸 브랜드에서 독자적으로 판매한 제품은 [11번가] 스팸클래식 340Gx2개+마일드340Gx2개 이것 하나임.
#판매가 매우 적음.
```

# 배송분석

```python
#평균 배송기간

print(naver['운송장등록소요일'].mean())
print(naver['출고소요일'].mean())
```

```python
#배송기간별 취소수량

print(naver.groupby('운송장등록소요일')['취소수량'].sum())
print(naver.groupby('출고소요일')['취소수량'].sum().sort_values(ascending=False))
```

```python
#배송기간별 반품수량

print(naver.groupby('운송장등록소요일')['반품수량'].sum().sort_values(ascending=False))
print(naver.groupby('출고소요일')['반품수량'].sum().sort_values(ascending=False))
#운송장등록에 1일 소요되거나 출고가 9일 소요된 경우에 반품수량이 가장 많았음.
```

```python
#배송이 평균보다 오래걸리는 상품(출고소요일로 비교)

naver[naver['출고소요일'] > 2]
```

```python
#문제의 반품수량이 가장 많았던 기간의 상품들

naver[(naver['운송장등록소요일'] == 1) | (naver['출고소요일'] == 9)]
```
